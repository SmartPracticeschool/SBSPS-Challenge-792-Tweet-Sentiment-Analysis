Making the tweets single line

```python
tweets=[]
for i in range(0, 1048574):
  r=[]
  r1=data['Tweets'][i]
  r=r1.split("(")
  del r[-1]
  tweets.extend(r)
```

```python
tweets
```

Removing special chars and converting to meaningful lines    


```python
import re

for i in range(len(tweets)):
  s=tweets[i]
  m=re.search("@",s)
  if(m):
    s=s[(m.end()+4):]
  
  s=re.sub("(n\'t)|(\w+nt)"," not",s)
  s=re.sub("(\'ve)"," have",s)
  s=re.sub("(\'s)"," is",s)
  s=re.sub("(\'m)"," am",s)
  s=re.sub("\\\\n","  ",s)
  s=re.sub(".\\\\.","  ",s)
  s=re.sub("\\\\.","  ",s)
  s=re.sub("\\\\x","  ",s)
  s=re.sub("x8","  ",s)
  s=re.sub('[^a-zA-Z]',' ',s)

  tweets[i]=s
```

```python
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
c=[]
for i in range(0,25337):
    tweet = re.sub('[^a-zA-Z]', ' ', sampleData['tweets'][i])
    tweet=tweet.lower()
    tweet=tweet.split()
    tweet = [word for word in tweet if not word in set(stopwords.words('english'))]
    
    #applystemming concept
    ps = PorterStemmer()
    tweet = [ps.stem(word) for word in tweet if not word in set(stopwords.words('english'))]

    tweet=' '.join(tweet)
    c.append(tweet)
```
